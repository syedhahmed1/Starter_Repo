## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The code cell below loads `data/raw/content_refresh_anonymized.csv` and prints counts that matter for a refresh-priority lane: how many pages exist, how many look visibly important, and how many are both visible and declining.

```python
import csv
from pathlib import Path

csv_path = Path("../../data/raw/content_refresh_anonymized.csv")
rows = 0
clients = set()
declining = 0
visible = 0
visible_declining = 0

with csv_path.open(newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        rows += 1
        clients.add(row["client_id"])
        impressions = float(row["impressions_90d"] or 0)
        is_declining = row["trend_direction"] == "down"
        if is_declining:
            declining += 1
        if impressions >= 500:
            visible += 1
            if is_declining:
                visible_declining += 1

print(f"Starter slice: {rows:,} pages across {len(clients)} clients")
print(f"Share with trend_direction == 'down': {declining / rows:.1%}")
print(f"Visible pages (impressions_90d >= 500): {visible:,}")
print(f"Visible AND declining (strong review candidates): {visible_declining:,}")
print()
print("From outputs/model_report.md on this same starter slice:")
print("  baseline_rules Precision@50 = 0.24")
print("  random_forest   Precision@50 = 0.74")
print("=> A learned ranker beats the hand rule at picking top-50 review targets (~3x lift).")
```
